4. Create an ordered list of HTML tags. Each list item must include the tag name and some information about
the tag.

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Tag</title>
</head>
<body>
    <ol>
        <li>
          <strong>&lt;html&gt;</strong>  The root element of an HTML document. It wraps all other elements except for the <code>DOCTYPE</code> declaration.
        </li>
        <li>
          <strong>&lt;head&gt;</strong>  Contains meta-information about the document, such as its title, character set, and links to stylesheets or scripts.
        </li>
        <li>
          <strong>&lt;title&gt;</strong>  Sets the title of the webpage, which appears in the browser tab.
        </li>
        <li>
          <strong>&lt;meta&gt;</strong>  Provides metadata such as character encoding, author, and viewport settings. It is a self-closing tag.
        </li>
        <li>
          <strong>&lt;link&gt;</strong>  Used to link external resources like stylesheets. This is also a self-closing tag.
        </li>
        <li>
          <strong>&lt;body&gt;</strong>  Contains the visible content of the web page, including text, images, and other elements.
        </li>
        <li>
          <strong>&lt;h1&gt; to &lt;h6&gt;</strong>  Heading tags that define titles and subheadings, where <code>&lt;h1&gt;</code> is the most important and <code>&lt;h6&gt;</code> the least.
        </li>
        <li>
          <strong>&lt;p&gt;</strong>  Represents a paragraph of text.
        </li>
        <li>
          <strong>&lt;a&gt;</strong>  Defines a hyperlink, which is used to link to another page or resource.
        </li>
        <li>
          <strong>&lt;img&gt;</strong>  Embeds an image into the document. This is a self-closing tag and requires a <code>src</code> attribute.
        </li>
        <li>
          <strong>&lt;ul&gt;</strong>  Creates an unordered list, typically with bullet points.
        </li>
        <li>
          <strong>&lt;ol&gt;</strong>  Creates an ordered list, typically with numbers or letters.
        </li>
        <li>
          <strong>&lt;li&gt;</strong>  Represents a list item within <code>&lt;ul&gt;</code> or <code>&lt;ol&gt;</code>.
        </li>
        <li>
          <strong>&lt;div&gt;</strong>  A generic container used for structuring and styling sections of a page.
        </li>
        <li>
          <strong>&lt;span&gt;</strong>  An inline container used to style a portion of text or group inline elements.
        </li>
      </ol>

</body>
</html>
```
